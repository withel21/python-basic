# 사전
## 개념
- 키와 값 쌍으로 데이터를 저장
- 중복된 키를 허용하지 않음
- 순서가 없음 -> 삽입된 순서와 무관하게 저장됨
## 생성
- 중괄호로 생성 - `{키1: 값1, 키2: 값2, ...}`
- `dict` 함수로 생성 - `dict([(키1, 값1), (키2, 값2), ...])`

```python
print('# Dictionary')
print('## creation')
x = {10: 'aaa', 20: 'bbb', 30: 'ccc'}
print(x)
x = [(10, 'aaa'), (20, 'bbb'), (30, 'ccc')]
y = dict(x)
print(y)
```

# 사전 연산
## 키 값으로 조회
- `사전명[키]`
## 키 추가/수정
- `사전명[키]=값`
## `in` 연산자 - 키 존재 여부 확인

```python
print('# Dictionary Operators')
print('## value by key')
x = {10: 'aaa', 20: 'bbb', 30: 'ccc'}
print(x[10])
x = {'aaa': 1, 'bbb': 2, 'ccc': 3}
x['ddd'] = 4
x['bbb'] = 5
print(x)
print()

print('## in operator')
x = {10: 'aaa', 20: 'bbb', 25: 'ccc'}
print(f'10 in {x}', 10 in x)
print(f"'aaa' in {x}", 'aaa' in x)
```

# 사전 내장 함수
## len: 사전내 요소의 개수를 알려주는 함수
- `len(d)`
## min/max: 사전 내 키 중 가장 작은/큰 값을 알려주는 함수
- `min(d)`/`max(d)`
## bool: 사전이 비어 있는지(`False` <- `{}`) 확인
- `bool(d)`

```python
print('# Dictionary internal functions')
print('## len')
x = {10: 'aaa', 20: 'bbb', 25: 'ccc'}
print(f'len({x})',len(x))
print()

print('## min/max')
print(f'min({x})',min(x))
print(f'max({x})',max(x))
print()

print('## bool')
print(f'bool({{}})',bool({}))
print(f"bool({{ 1:'aaa', 2:'bbb'}})", bool({1:'aaa', 2:'bbb'}))
```

# 사전 함수
- `keys()` : 모든 키를 반환
- `values()` : 모든 값을 반환
- `items()` : 모든 키-값 쌍을 튜플 형태로 반환
- `get(key)`: `key`에 해당 하는 값을 반환(없으면 `None`) 
- `clear()` : 사전에 저장된 모든 항목을 삭제
- `pop(key)` : 사전에서 `key`에 해당하는 항목을 삭제하고, 값을 반환

```python
print('# Dictionary functions')
print('## keys')
x = {10: 'aaa', 20: 'bbb', 25: 'ccc'}
print(f'{x}.keys()\n', x.keys())
print()

print('## values')
print(f'{x}.values()\n', x.values())
print()

print('## items')
print(f'{x}.items()\n', x.items())
print()

print('## get')
print(f'{x}.get(10)', x.get(10))
print(f'{x}.get(30)', x.get(30))
print()

print('## pop')
print(f'{x}.pop(20)', x.pop(20))
print(x)
print()

print('## clear')
x.clear()
print(x)
print()

```
