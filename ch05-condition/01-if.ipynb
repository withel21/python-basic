# if 문

## 기본 if 문
- 단수 if 문
```
if 조건:
  # 조건이 참일 때 실행할 코드
```
- if ~ else 문
```
if 조건:
  # 조건이 참일 때 실행할 코드
else:
  # 조건이 거짓일 때 실행할 코드
```
- if ~ elif ~ else 문
```
if 조건1:
  # 조건1이 참일 때 실행할 코드
elif 조건2:
  # 조건2가 참일 때 실행할 코드
else:
  # 모든 조건이 거짓일 때 실행할 코드
```
- 중첩 if문
```
if 조건1:
  # 조건1이 참일 때 실행할 코드
  
  if 조건2:
    # 조건1과 조건2가 참일 때 실행할 코드
  else:
    # 조건1이 참이고 조건2가 거짓일 때 실행할 코드
else:
  # 조건1이 거짓일 때 실행할 코드
```

````python
print('# if statement')
print('## if basic')
x = 95
if x >= 60:
  print('x는 60이상')
  print('{0}는 60 이상'.format(x))
print()

print('## if~else')
x = 35
if x >= 60:
  print('x는 60이상')
  print('{0}는 60 이상'.format(x))
else:
  print('x는 60미만')
  print('{0}는 60 미만'.format(x))
print()

print('## if ~ elif ~ else')
x = 79
if x >= 90:
  print('x는 90이상')
elif x >= 80:
  print('x는 80이상')
else:
  print('x는 80미만')
print()

print('## nested if stmt')
x, y = 30, 40
if x >= 60:
  print('x는 60 이상이고, ')
  if y >= 60:
    print('y는 60 이상')
  else:
    print('y는 60 미만')
else:
  print('x는 60 미만이고, ')
  if y >= 60:
    print('y는 60 이상')
  else:
    print('y는 60 미만')
````

## if 문과 함께 사용하는 연산자
- or 연산자: 하나만 참이면 참
- and 연산자: 모두 참이어야 참
- not 연산자: 참 -> 거짓, 거짓 -> 참으로 반대로 뒤집음
- in 연산자: 주어진 값이 특정한 자료형에 포함되어 있는지 확인

````python
print('# operator in if stmt')
print('# or')
x, y = 90, 40
if x >= 60 or y >= 60:
  print('x는 60 이상이거나, y는 60 이상')
else:
  print('x는 60 이상도 아니고, y도 60 이상이 아님')
print()

print('# and')
if x >= 60 and y >= 60:
  print('x는 60 이상이면서 y는 60 이상')
else:
  print('x는 60 이상이 아니거나, y는 60 이상이 아님')
print()

print('# not')
x = 55
if not x >= 60:
  print('x는 60 이상이 아님')
else:
  print('x는 60 이상')
print()

print('# in (string)')
x = 'Hello'
if 'h' in x:
  print('h 있음')
elif 'H' in x:
  print('H는 있음')
else:
  print('h, H 둘 다 없음')

print('# in (list)')
x = [10, 20, 30]

if 15 in x:
  print('15는 있음')
elif 20 not in x:
  print('20은 없음')
else:
  print('15는 없고, 20은 있음')

print('# in (set, dictionary)')
x, y = {10, 20, 30}, {10: 'aaa', 20: 'bbb', 30: 'ccc'}

if 15 in x or 25 in y:
  print('x에 15가 있거나, y에 25 가 있음')
elif 20 in x and 25 not in y:
  print('x에 20이 있고, y에 25가 없음')
else:
  print('if문, elif문 둘 다 거짓')
````
