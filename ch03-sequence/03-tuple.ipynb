# 튜플
## 개념 - 변경할 수 없는 자료형
> 리스트와 비교해서, `[]` 대신 `()`를 사용하고, 수정, 삭제를 할 수 없다는 점이 다름
## 생성
- `튜플명 = (요소1,)` : 하나의 요소만 있는 경우
> 반드시 `,`를 넣어야 요소 하나의 튜플로 인식됨
- `튜플명 = (요소1,요소2,...)` : 여러개의 요소가 있는 경우
## 인덱싱/슬라이싱
- 리스트와 동일

```python
print('# Tuple')
print('## creation')
a = (1, 2, 3, True, False, 'ABC', 'DEF')
print(a)
print()
print('## indexing/slicing')
a = (10, 20, 30, 40, 50)
print(a[0], a[1], a[2], a[3], a[4])
print(a[-5], a[-4], a[-3], a[-2], a[-1])
print(a[1:4:2])
```
