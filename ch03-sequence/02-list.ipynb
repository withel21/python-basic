# 리스트
## 개념 - 크기가 가변적으로 변하는 선형 리스트
## 생성
- `리스트명 = [요소1, 요소2, ...]`
- `리스트명 = list((요소1, 요소2, ...))`
## 인덱싱
- 리스트의 길이가 n일때
  - 첫번째 요소 : 0, -n
  - 두번째 요소 : 1, -(n-1)
  - ...
  - 뒤에서 두번째 요소: n-2, -2
  - 뒤에서 첫번째 요소: n-1, -1


```python
print('# List')
print('## creation')
a = [1, 2, 3, True, False, 'ABC', 'DEF']
print(a)
b = list((1, 2, 3, True, False, 'ABC', 'DEF'))
print(b)
print()

print('## indexing')
a = [10, 20, 30, 40, 50]
print(a[0], a[1], a[2], a[3], a[4])
print(a[-5], a[-4], a[-3], a[-2], a[-1])
print()
```

## 슬라이싱
### 형태 - `list[start:stop:step]`
#### 순방향 - `step`이 없거나 양수
- `start` - 슬라이싱이 시작되는 인덱스(생략 -> 0)
- `stop` - 슬라이싱이 끝나는 인덱스(생략 -> 문자열 끝)
- `step` - 건너뛸 칸 수 (생략 -> 1: 하나씩 증가하면서 잘라냄)
#### 역방향 - `step`이 음수일 때
- `start` - 슬라이싱이 시작되는 인덱스(생략 -> 문자열 끝)
- `stop` - 슬라이싱이 끝나는 인덱스(생략 -> 0)
- `step` - 건너뛸 칸 수


```python
print('## List slicing')
a = [10, 20, 30, 40, 50, 60, 70, 80, 90]
print(a[1:8:2])
print(a[1:8:])
print(a[1::2])
print(a[1::])
print(a[:8:2])
print(a[:8:])
print(a[::2])
print(a[::])
print(a[-2:-5:-1])
print(a[-2::-1])
print(a[:-5:-1])
print(a[::-1])
print()
```

## 수정
- 한개 수정 : `list[index]=value`
- 여러개 수정: `list[start:end]=new_values` - `start` 에서 `end` 이전 까지의 인덱스 내용을 `new_values`로 대체
## 삭제
- 한개 삭제 : `del list[index]`
- 여러개 삭제 : `del list[start:end]`

```python
print('## List update')
a = [1, 2, 3, 4, 5]
a[1] = 9
print(a)
a[1:4] = [6, 7, 8]
print(a)
a[1:3] = [2, 3, 4, 5]
print(a)
print()

print('## List delete')
a = [1, 2, 3, 4, 5]
del a[1]
print(a)
a = [1, 2, 3, 4, 5]
del a[1:4]
print(a)
```

# 리스트와 함께 사용하는 연산자
## 숫자 연산자
- `+` : 리스트 연결 연산자
- `*` : 리스트 반복 연산자
## 비교 연산자
- `==` : 두 리스트가 같은가? 
- `!=` : 두 리스트가 다른가? 
- `<` : 왼쪽 리스트가 오른쪽 리스트보다 작은가?
- `>` : 왼쪽 리스트가 오른쪽 리스트보다 큰가?
- `<=` : 왼쪽 리스트가 오른쪽 리스트보다 같거나 작은가?
- `>=` : 왼쪽 리스트가 오른쪽 리스트보다 같거나 큰가?
## 대입 연산자/복합 대입 연산자
- `=` : 왼쪽의 변수에 오른쪽의 리스트를 대입
- `+=` : 왼쪽의 변수에 왼쪽의 리스트에 오른쪽의 리스트를 붙여서 대입
- `*=` : 왼쪽의 변수에 오른쪽의 숫자 만큼 왼쪽의 리스트를 반복해서 대입
## `in` 연산자
- `값 in 리스트` : 값이 리스트안에 있는가?

```python
print('## List concat/repeatance operator')
a = [10, 20, 30, 40]
b = ['ab', 'cd', 'ef', 'gh']
print(a+b)
print(a*3)
print()

print('## List comparison operator')
a = [1, 2, 3]
b = [1, 2, 4]

print(f'{a} == {b} ?', a == b)
print(f'{a} != {b} ?', a != b)
print(f'{a} < {b} ?', a < b)
print(f'{a} <= {b} ?', a <= b)
print()

print('## List assignment operator')
a = [10, 20, 30]
b = [15, 25, 35]
a += b
print(a)
a *= 2
print(a)
print()

print('## List in operator')
a = [10, 20, 30]
print(10 in a)
print(15 in a)
print(10 not in a)
print(15 not in a)
```

# 리스트와 함께 사용하는 내장 함수
## len : 리스트 내의 요소 개수를 알려주는 함수
- `len(l)`
## min/max : 리스트에서 가장 작은/큰 값을 알려주는 함수
- `min(l)`/`max(l)`
## bool : 리스트가 비어있는지(`False`) 확인
- `bool(l)`

```python
print('## List internal functions')
print('### len')
a = [10, 30, 20, 50, 40]
print(f'len({a})', len(a))
print()

print('### min/max')
print(f'min({a})',min(a))
print(f'max({a})',max(a))
print()

print('### bool')
print(f'bool([])', bool([]))
print(f'bool([1, 2])', bool([1, 2]))

```

## 리스트 함수
### count : 리스트내에서 특정 값의 개수를 알려주는 함수
- `list.count(value)`

### index : 리스트에서 해당 요소가 처음 등장하는 index를 알려주는 함수
- `list.index(value, start, end)`
  - `value`(필수) - `list`에서 찾을 값
    > 주의: 리스트에 없는 값을 넣으면 에러! 
  - `start`(선택) - 검색 시작 인덱스(없으면 0)
  - `end`(선택) - 검색 끝 인덱스(없으면 `len(list)`)

### append : 리스트 마지막 요소 뒤에 값을 추가
- `list.append(value)`

### extend : 리스트의 끝에 시퀀스 객체(리스트, 문자열, 튜플)을 추가
- `list.extend(value)`

### insert : 리스트의 특정 위치에 요소를 삽입
- `list.insert(index,value)`
  - `index` : 요소를 삽입할 인덱스, 이 인덱스 값에 요소가 삽입되고, 그 뒤에 있는 요소들은 하나씩 밀림
  - `value` : 삽입할 값

### copy : 리스트를 복사하는 함수
- `list.copy()`

### remove : 리스트에서 첫번째로 나타나는 특정 요소를 제거
- `list.remove(value)`
  > 주의: 리스트에 없는 값을 넣으면 에러! 

### pop : 리스트에서 특정 인덱스의 요소를 제거하고 그 값을 반환
- `list.pop(index)` : `index`를 지정하지 않으면, 마지막 요소를 제거하고 그 값을 반환
  > 주의: 빈 리스트에서 이 함수을 부르면 에러! 

### reverse : 리스트의 요소를 역순으로 재배열
- `list.reverse()`

### sort : 리스트의 항목을 기준에 따라 정렬
- `list.sort(key,reverse)`
  - `key`(선택) - 정렬기준을 정의 하는 함수(기본값: `None`)
  - `reverse`(선택) - 정렬방식
    - `True`: 내림차순
    - `False`: 오름차순(기본값)

### clear: 리스트의 모든 요소를 제거하여 빈 리스트(`[]`)로 만듦
- `list.clear()`

```python
print('# List functions')
print('## count')
a = [10, 20, 30, 10, 20, 30]
print(f'{a}.count(10)', a.count(10))
print(f'{a}.count(5)', a.count(5))
print()

print('## index')
print(f'{a}.index(20)', a.index(20))
print(f'{a}.index(30,3)', a.index(30,3))
print()

print('## append')
a = [1, 2, 3]
a.append(4)
print(a)
print()

print('## extend')
a = [1, 2, 3]
b = [4, 5, 6]
b.extend(a)
print(b)
a.extend('abc')
print(a)
print()

print('## insert')
a = [10, 20, 30]
a.insert(1, 15)
print(a)
```

```python
print('## copy')
a = [1, 2, 3]
b = a.copy()
b[0] = 10
print(a)
print(b)
print()

print('## remove')
a = [1, 2, 3, 2, 4]
a.remove(2)
print(a)

print('## pop')
a = [10, 20, 30, 40, 50]
print(f'{a}.pop()', a.pop())
print(a)
print(f'{a}.pop(1)', a.pop(1))
print(a)
print()

print('## reverse')
a = [1, 2, 3, 4, 5]
a.reverse()
print(a)
print()

print('## sort')
a = [3, 1, 4, 1, 5, 9, 2]
a.sort(reverse=True)
print(a)
b = ['cherry', 'banana', 'apple', 'durian']
b.sort(key=len)
print(b)
print()

print('## clear')
a = [1, 2, 3, 4, 5]
a.clear()
print(a)
```

## 다차원 리스트
- 2차원 이상의 리스트로 `[` 와 `]` 사이에 `[` 와 `]` 를 중첩하여 사용

```python
print('# multi dimension List')
a = [[1, 2], [3, 4], [5, 6]]
print('a', a)
print('a[0]', a[0])
print('a[1][0]', a[1][0])
b = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]
print('b', b)
print('b[0]', b[0])
```
