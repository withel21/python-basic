```python
print(True == True)
print(True == False)
print(True != True)
print(True != False)
print(True < True)
print(True < False)
print(True <= True)
print(True <= False)
```

## 논리연산자

```python
print('and operator')
print(True and True)
print(True and False)
print(False and True)
print(False and False)

print('or operator')
print(True or True)
print(True or False)
print(False or True)
print(False or False)

print('not operator')
print(not True)
print(not False)
```
