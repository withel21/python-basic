# for 문

## 문법
```
for 변수 in 반복가능한객체:
  # 실행할 코드
```
- 반복가능한 객체
  - 리스트
  - 문자열
  - 집합
  - 사전 - 키를 접근


````python
print('## for with list')
a = [10, 20, 30, 40, 50]
for x in a:
  print(x)

print()
print('## for with sliced list')
for x in a[1:4:2]:
  print(x)

print()
print('## for with string')
a = 'Hello'
for x in a:
  print(x)

print()
print('## for with set')
a = {10, 20, 30, 40, 50}
for x in a:
  print(x)

print()
print('## for with dictionary')
a = {10: 'aaa', 20: 'bbb', 30: 'ccc'}
for x in a:
  print(x, a[x])
````

## 리스트 내포
- 조건이 없는 경우
```
[표현식 for 항목 in 반복가능한 객체]
```
- 조건이 포함된 경우
```
[표현식 for 항목 in 반복가능한 객체 if 조건]
```

````python
print('## list with inclusive for statement(no condition)')
a = [1, 2, 3, 4, 5]
b = [num ** 2 for num in a]
print(b)

print()
print('## list with inclusive for statement(with condition)')
b = [num ** 2 for num in a if num % 2 == 1]
print(b)
````

## for문과 함께 사용하는 함수

### `range` - 지정된 범위의 숫자 시퀀스를 만들어 내는 함수
```
range(start, stop, step)
```
- start: 시작할 숫자(default:0)
- stop: 종료 값(포함되지 않음)
- step: 숫자 사이의 간격(default: 1)

### `zip` - 여러 반복 가능한객체(리스트, 튜플 등)을 묶어서 동시에 접근하게 하는 함수
- 서로 길이가 다르면 길이가 짧은 객체에 맞춰서 반복

````python
print('## range function')
print(list(range(1,8,2)))
print(list(range(1,8)))
print(list(range(8)))
print()

print('## for with range')
for j in range(1,10):
  print(f"{2} X {j} = {2 * j}")
print()

print('## for with zip')
a = [1,2,3,4]
b = [4,5,6]
for x, y in zip(a, b):
  print(x, y)
````

## for문과 if문 + 중첩 for문

````python
print('## if in for')
a = [1,2,3,4,5]
for x in a:
  if x % 2 == 0:
    print(f"{x}는 짝수")

print()

print('## nested for - 2d list')
x = [[1,2], [3,4]]
for i in x:
  for j in i:
    print(j)
print()

print('## nested for')
for i in range(2, 10):
  for j in range(1, 10):
    print(f"{i} X {j} = {i*j}")
  print()
````
